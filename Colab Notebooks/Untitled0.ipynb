```python
import requests
from bs4 import BeautifulSoup

```

```python
Response = requests.get('https://www.google.com/')
```

```python
content = Response.text
```

```python
soup = BeautifulSoup(content, 'html.parser')
data = soup.find_all('a')

for i in data:
    print(i.get_text())
```
