```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in data.columns:
  data[col] = le.fit_transform(data[col])
```

```python
data['Attendance*Previous_Scores']=data['Attendance'].values*data['Previous_Scores'].values
data['Hours_Studied*Previous_Scores']=data['Hours_Studied'].values*data['Previous_Scores'].values
data['Hours_Studied*Attendance']=data['Hours_Studied'].values*data['Attendance'].values
data['Sleep_Hours*Previous_Scores']=data['Sleep_Hours'].values*data['Previous_Scores'].values
data['Attendance*Previous_Scores']=data['Attendance'].values*data['Previous_Scores'].values


data['Attendance**2']=data['Attendance']**2
data['Previous_Scores**2']=data['Previous_Scores']**2
data['Hours_Studied**2']=data['Hours_Studied']**2
data['Sleep_Hours**2']=data['Sleep_Hours']**2

data['Attendance/2']=data['Attendance']**2
data['Previous_Scores/2']=data['Previous_Scores']/2
data['Hours_Studied/2']=data['Hours_Studied']/2
data['Sleep_Hours/2']=data['Sleep_Hours']/2


data
```
