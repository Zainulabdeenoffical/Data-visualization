```python
import pandas as pd
file_path = '/content/drive/MyDrive/CSV/ai_job_market_insights.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
# Check for missing values
print("Missing values per column:\n", df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the salary distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['Salary_USD'], bins=20, kde=True)
plt.title('Salary Distribution')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
df_encoded = pd.get_dummies(df, drop_first=True)

# Bar plot of average salary per industry
plt.figure(figsize=(10, 6))
sns.barplot(x='Industry', y='Salary_USD', data=df, estimator=lambda x: sum(x) / len(x))
plt.title('Average Salary by Industry')
plt.xticks(rotation=45)
plt.ylabel('Average Salary (USD)')
plt.show()
```

```python
# Boxplot of AI Adoption Level vs Salary
plt.figure(figsize=(8, 5))
sns.boxplot(x='AI_Adoption_Level', y='Salary_USD', data=df)
plt.title('AI Adoption Level vs Salary')
plt.show()
```

```python
# Select only numeric columns for the heatmap
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_columns.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Numeric Features)')
plt.show()
```
