```python
import plotly.express as px

fig = px.line(df, x='Year', y='GDP (in Trillions USD)', color='Country', title='GDP Over the Years by Country')
fig.show()
```

```python
fig = px.bar(df, x='Country', y='GDP per Capita (in USD)', color='Year', title='GDP per Capita Comparison by Country')
fig.show()
```

```python
fig = px.pie(df, names='Country', values='Renewable Energy Share (%)', title='Renewable Energy Share by Country')
fig.show()
```

```python
fig = px.histogram(df, x='Poverty Rate (%)', color='Country', nbins=20, title='Poverty Rate Distribution Across Countries')
fig.show()
```
